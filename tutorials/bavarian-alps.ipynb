```python
import os
import plotly.io as pio


from stravalib import Client

from strava_collections.collection import StravaCollection

%load_ext autoreload
%autoreload 2
```

```python
pio.renderers.default = "notebook"
```

```python
client_id = os.getenv("STRAVA_CLIENT_ID")
client_secret = os.getenv("STRAVA_CLIENT_SECRET")
refresh_token = os.getenv("STRAVA_REFRESH_TOKEN")

client = Client()

# Refresh access token
token_response = client.refresh_access_token(
    client_id=client_id,
    client_secret=client_secret,
    refresh_token=refresh_token,
)
client.access_token = token_response["access_token"]
```

```python
bavarian_alps = [
    (12284003055, True),
    # (12231765525, False),
    # (11603714471, True),
    # (11487681160, False),
    # (11492585278, False),
    # (9020068072, True),
    # (11518973205, False),
    # (9327605554, False),
    # (9586696097, True),
    # (9961995244, False),
    # (11166740981, True),
    # (11378142289, True),
    # (11368440756, False),
    # (15177143911, False),
    # (15877224330, False),
]
activity_ids = bavarian_alps

# Create collection and plot
collection = StravaCollection(client, activity_ids=activity_ids)
```

```python
fig_elev = collection.plot_elevation()
fig_elev.show()
```

```python
fig_map = collection.plot_map()
fig_map.show()
```

```python
tot_elevation_gaion = 0.0
for activity in collection.activities:
    tot_elevation_gaion += activity.activity.total_elevation_gain
    print(
        f"Activity: {activity.activity.name}, elevation gain: {activity.activity.total_elevation_gain}"
    )
print(f"{tot_elevation_gaion = }")
```
